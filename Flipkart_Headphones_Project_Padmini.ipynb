```python
# Extract Brand
df1['Brand'] = df1['Product Name'].str.extract(r'^(\w+(?: \w+)?)')
```

```python
# Extract Colour
df1['Colour'] = df1['Features'].str.extract(r'(?i)(Black|Green|Beige|Blue|Red|White|Yellow|Grey|Brown|Pink|Purple|Orange)')
```

```python
import re

# Regex-based function to extract mode
def extract_mode_regex(product_name):
    match = re.search(r'(Wireless|Bluetooth|Wired)', product_name, re.IGNORECASE)
    return match.group(1).capitalize() if match else "Unknown"

# Apply the regex function to create a new column in df1
df1["Mode"] = df1["Product Name"].apply(extract_mode_regex)

# Display the updated table
df1
```

```python
df1
```

```python
# Extract Model
df1['Model'] = df1['Features'].str.extract(r'(?i)(True Wireless|On the Ear|In the Ear)')


```

```python
df1
```

```python
# Extract Playtime from Product Name
df1['Playtime'] = df1['Product Name'].str.extract(r'(\d+\s*(?:hrs|hours|h|HRS|Hrs|H))', expand=False)

```

```python
# Function to extract the version from product_name
def extract_version(name):
    # Example: look for known keywords for versioning
    match = re.search(r'(Hybrid|Elite|Supreme|Pro|Z2|Vs102|110|42)', name, re.IGNORECASE)
    return match.group(1) if match else "Standard"

# Create the new 'version' column
df1['version'] = df1['Product Name'].apply(extract_version)
```

```python
# Assuming df1 is your DataFrame
df1 = df1.loc[:, ~df1.columns.str.contains('^unnamed:_0')]
```

```python
##DATA CLEANING
```

```python
# Step 1: Rename columns for consistency
df1.columns = df1.columns.str.strip().str.replace(' ', '_').str.lower()
```

```python
# Step 2: Clean 'price' column (remove currency symbol and convert to numeric)
df1['price'] = df1['price'].str.replace('', '').str.replace(',', '').astype(float)
```

```python
df1['rating'] = df1['rating'].astype(float)
```

```python
# Clean 'playtime' column
df1['playtime'] = df1['playtime'].astype(str).str.extract(r'(\d+)').astype(float)

# Handle missing values by replacing NaN with 0
df1['playtime'] = df1['playtime'].fillna(0)

```

```python
df1 = df1.dropna()  # Drop rows with missing values in other columns
```

```python
# Step 5: Remove duplicates (if any)
df1 = df1.drop_duplicates()
```

```python
# Step 6: Standardize text columns (e.g., brand and color)
df1['product_name'] = df1['product_name'].str.strip().str.title()
df1['brand'] = df1['brand'].str.strip().str.title()  # Capitalize each word in 'brand'
df1['colour'] = df1['colour'].str.strip().str.title()  # Capitalize each word in 'colour'
```

```python
# Step 8: Verify logical values
df1 = df1[(df1['rating'] >= 0) & (df1['rating'] <= 5)]  # Ensure ratings are between 0 and 5

# Replace 'Unknown' with 'Wired' in the 'mode' column
df1['mode'] = df1['mode'].replace('Unknown', 'Wired')

# Print the updated DataFrame
print(df1)
```

```python
# Drop the 'product_name' column
df1.drop('product_name', axis=1, inplace=True)
df1.drop('features', axis=1, inplace=True)

print(df1)
```

```python
# Replace 'Unknown' with 'Wired' in the 'mode' column
df1['mode'] = df1['mode'].replace('Wireless', 'Bluetooth')
```
